```python
from collections import deque

```

```python

# State = (farmer, wolf, goat, cabbage)
# Each value is 'W' (West) or 'E' (East)

INITIAL = ('W', 'W', 'W', 'W')
GOAL    = ('E', 'E', 'E', 'E')

```

```python


def is_safe(state):
    farmer, wolf, goat, cabbage = state
    # Wolf eats goat if left alone
    if wolf == goat and wolf != farmer:
        return False
    # Goat eats cabbage if left alone
    if goat == cabbage and goat != farmer:
        return False
    return True

```

```python

def get_next_states(state):
    farmer, wolf, goat, cabbage = state
    new_side = 'E' if farmer == 'W' else 'W'
    next_states = []

    # Farmer crosses alone
    s = (new_side, wolf, goat, cabbage)
    if is_safe(s):
        next_states.append((s, 'Farmer crosses alone'))

    # Farmer takes wolf
    if wolf == farmer:
        s = (new_side, new_side, goat, cabbage)
        if is_safe(s):
            next_states.append((s, 'Farmer takes Wolf'))

    # Farmer takes goat
    if goat == farmer:
        s = (new_side, wolf, new_side, cabbage)
        if is_safe(s):
            next_states.append((s, 'Farmer takes Goat'))

    # Farmer takes cabbage
    if cabbage == farmer:
        s = (new_side, wolf, goat, new_side)
        if is_safe(s):
            next_states.append((s, 'Farmer takes Cabbage'))

    return next_states

```

```python



def solve():
    queue = deque()
    queue.append((INITIAL, [('Start', INITIAL)]))
    visited = set([INITIAL])

    while queue:
        state, path = queue.popleft()

        if state == GOAL:
            return path

        for next_state, action in get_next_states(state):
            if next_state not in visited:
                visited.add(next_state)
                queue.append((next_state, path + [(action, next_state)]))

    return None

# Run and print solution
solution = solve()

print('Goat, Wolf and Cabbage - Solution')
print('State format: (Farmer, Wolf, Goat, Cabbage)')
print('-' * 45)

for i, (action, state) in enumerate(solution):
    print(f'Step {i}: {action}')
    print(f'        State: {state}')
    print()

print(f'Total moves: {len(solution) - 1}')
```
