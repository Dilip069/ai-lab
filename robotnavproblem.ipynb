```python
from collections import deque
grid =[
    ['s', ' ', ' ', 'X ', ' '],
    [' ', 'X', ' ', 'X', ' '],
    [' ', 'X', ' ', ' ', ' '],
    [' ', ' ', 'X', 'X', ' '],
    [' ', ' ', ' ', ' ', 'g']
]
print("Grid layout:")
for row in grid:
    print(row)
```

```python
initial_state = (0, 0)  # Starting position of 's'
goal_state = (4, 4)     # Position of 'g'
actions = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # right, down, left, up
def is_valid(row,col):
    return 0 <= row < 5 and 0 <= col < 5 and grid[row][col]== 0

    
```

```python
def bfs():
    queue = deque()
    queue.append((initial_state, [initial_state]))
    visited = set([initial_state])

    while queue:
        (row, col), path = queue.popleft()

        if (row, col) == goal_state:
            return path

        for dr, dc, name in actions:
            nr, nc = row + dr, col + dc
            if is_valid(nr, nc) and (nr, nc) not in visited:
                visited.add((nr, nc))
                queue.append(((nr, nc), path + [(nr, nc)]))

    return None
```
