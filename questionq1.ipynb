```python

def is_safe(board, row, col, n):
    # column test
    for x in range(row):
        if board[x][col] == 'Q':
            return False

    # diagonal test
    # top-left diagonal test
    r = row - 1
    c = col - 1
    while r >= 0 and c >= 0:
        if board[r][c] == 'Q':
            return False
        r -= 1
        c -= 1

    # top-right diagonal test
    r = row - 1
    c = col + 1
    while r >= 0 and c < n:
        if board[r][c] == 'Q':
            return False
        r -= 1
        c += 1

    return True



    
```

```python
def nQueenSolver(board,x,n):
    if x>= n:
        return True
    for col in range(x):
        if is_safe(board,x,col,n):
            board[x,col]= 'Q'
            if nQueenSolver(board,x+1,n):
                return True
            board[x][col]='_'
            
    return False        
```

```python
if nQueenSolver(board,0,n):
    pprint(board)
```
